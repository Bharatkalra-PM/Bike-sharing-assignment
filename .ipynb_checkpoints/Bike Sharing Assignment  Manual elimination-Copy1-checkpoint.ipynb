## Step 2: Cleaning the dataset. 
        - Applying imputations and dropping irrelevant columns

```python
bikesharing.head()
```

```python
#Dropping below columns:
    # - dteday since we wont do day-to day analysis here. ALso, we already have month, year and weekday columns separately
    # - instant since it is just an index and nothing else
    # - casual and registered since we have cnt as the target variable which is a sum of casual and registered
    
bikesharing.drop(['registered', 'casual', 'dteday', 'instant'], axis=1, inplace= True)
```

```python
# Look at
# Statistical Significance(p_values)
# VIF(Variance inflation factor) for checking Multicollinearity

# Based on these 2, take a call to decide which feature/variable to be removed.
```

```python
# VIF(i) = 1/(1-(r(i))^2)
```

```python
#Create a dataframe that will contain the names of all the feature variables and their respecive

vif = pd.DataFrame()
vif['Features'] = X_train.columns
vif['VIF'] = [variance_inflation_factor(X_train.values,i) for i in range(X_train.shape[1])]
vif['VIF'] = round(vif['VIF'],2)
vif = vif.sort_values(by = 'VIF', ascending = False )
vif
```

#### We can have below scenarios
    - High p-value, High VIF //No brainers
    - High - low :
        - High p, low VIF //Remove these first
        - Low p , High VIF  //Remove these after
    - Low p, Low VIF

# Step 5: Prediction and Evaluation on Test set

```python
#For this, we compute R^2 for test set and then compare R^2 values for train and test set.
```

```python
#Important step - The coefficients that the model has learnt correspond to the scaled features. 
#So, we have to do the same pre processing transformation steps to test data set as well.
```

```python
#fit() : learns xmin, xmax  --> ON TRAINING DATA SET

#transform() : x-xmin/xmax-xmin  --> ON TEST DATA SET 

#fit_transform() : learns and transforms simultaneously

#We can't fit the scaler on test data set because test data set is unseen
```

```python
# Create a list of only numeric variables
num_vars = ['temp', 'atemp','hum','windspeed','cnt']

# Fit on data
df_test[num_vars] = scaler.transform(df_test[num_vars])
df_test.head()
```

```python
df_test.describe()
```

#### Now good to go with predictions

```python
#X_test, y_test

y_test = df_test.pop('cnt')
X_test = df_test 
```

```python
#Add a constant to X_test to make predictions
#Nuance - Add the constant as statsmodel doesnt add that
X_test_sm = sm.add_constant(X_test)
X_test_sm.head(10)
```

```python
#Remove the variables that were removed from training model

X_test_sm = X_test_sm.drop(['oct','atemp','holiday','sun','spring','jul','jun','mar','wed','thu','may','tue'],axis=1)
```

```python
#Predict
y_test_pred = lr_model.predict(X_test_sm)
```

```python
#Evaluate
r2_score(y_true = y_test, y_pred = y_test_pred)
```
